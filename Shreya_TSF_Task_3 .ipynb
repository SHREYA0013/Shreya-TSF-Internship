#### Univariate Analysis

Shipping Mode:

```python
df["Ship Mode"].value_counts() #shipping mode wise
```

```python
ax=sns.countplot(x='Ship Mode',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
plt.show() #creating histograms
```

```python
df['Ship Mode'].value_counts().plot(kind='pie',figsize=[5,5],autopct='%1.1f%%')
plt.title('Pie chart order distribution according to Shipping Mode')
```

City:


```python
df['City'].value_counts()
```

```python
fig, ax = plt.subplots(figsize=(15, 6))
ax = sns.countplot(x='City', data=df, order=df['City'].value_counts().index[:10])
for bars in ax.containers:
    ax.bar_label(bars)
    print("Mostly ordered from")
plt.show()
```

##### New York city has the most orders followed by Los Angeles

```python

fig, ax = plt.subplots(figsize=(15, 6))
ax = sns.countplot(x='City', data=df, order=df['City'].value_counts().index[-10:])
for bars in ax.patches:
    ax.annotate(f'{bars.get_height()}', (bars.get_x() + bars.get_width() / 2., bars.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')
print("Least order placed from")
plt.show()
```

### Bivariate Analysis

```python
df.columns
```

```python
regional_analysis=df.groupby("Region").sum(numeric_only=True)
print(regional_analysis)
```

```python
regional_analysis[['Sales','Profit']].plot.bar()
```

```python
categorical_analysis=df.groupby("Category").sum(numeric_only=True)
print(categorical_analysis)
```

```python
categorical_analysis[['Sales','Profit']].plot.bar()
```

```python
city_analysis=df.groupby("City").sum(numeric_only=True)
print(city_analysis)
```

```python

top_10_cities = city_analysis.nlargest(10, 'Sales')

top_10_cities[['Sales', 'Profit']].plot.bar()

plt.title('Top 10 Cities: Sales and Profit Analysis')
plt.xlabel('City')
plt.ylabel('Amount')
plt.show()
```

some of the cities have negative profit too
